````python
from pathlib import Path
import sys
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
````

````python

# Détecte le répertoire racine du dépôt.
root = Path.cwd()
if root.name == "notebook":
    root = root.parent

# Assure l'accès au module src.
sys.path.insert(0, str(root / "src"))

from dataset_loader import load_brain_tumor_dataset

DATASET_PATH = root / "data" / "archive" / "Data"
print(f"Dataset path: {DATASET_PATH}")

images, labels = load_brain_tumor_dataset(str(DATASET_PATH))
print("images shape:", images.shape)
print("labels shape:", labels.shape)
````

## Exemples d'images originales
Affichage de quelques images du dataset après conversion en niveaux de gris et redimensionnement.

````python
# Affiche 5 images originales provenant de différentes classes.
num_examples = 5
indices = np.linspace(0, len(images) - 1, num_examples, dtype=int)
fig, axes = plt.subplots(1, num_examples, figsize=(15, 4))

for ax, idx in zip(axes, indices):
    image = images[idx].squeeze()
    label = labels[idx]
    ax.imshow(image, cmap="gray")
    ax.set_title(f"Label: {label}")
    ax.axis("off")

plt.suptitle("Exemples d'images originales du dataset")
plt.show()
````
